```python
df = pd.read_csv('../model_output/in.csv', index_col=0)
```

```python
final_list = []
for i in range(len(df)):
    pred_labels = df['predicted_label'][i][1:-1].split(',')
    pred_boxes = df['predicted_box'][i][2:-1].split('],')

    for ii in range(len(pred_labels)):
        id = df['id'][i]
        label = pred_labels[ii]
        pred_boxes[ii] = pred_boxes[ii].split(',')
        x0 = int(pred_boxes[ii][0].split('.')[0].replace('[',''))
        y0 = int(pred_boxes[ii][1].split('.')[0].replace('[',''))
        x1 = int(pred_boxes[ii][2].split('.')[0].replace('[',''))
        y1 = int(pred_boxes[ii][3].split('.')[0].replace('[',''))

        final_list.append(id, label, x0, y0, x1, y1)

    break
```

```python
# save list to tsv file
def save_list_to_tsv_file(path, output_list):
    with open(path, 'w', newline='') as f_output: 
        tsv_output = csv.writer(f_output, delimiter='\n')
        tsv_output.writerow(output_list)
```
