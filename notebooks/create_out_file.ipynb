```python
import pandas as pd
import csv
import sys
sys.path.append('../src/')
from functions import from_tsv_to_list
```

```python
df = pd.read_csv('../model_output/model_output.csv', index_col=0)
```

```python
df
```

```python
name_list, out_list = [], []
for i in range(len(df)):
    pred_labels = pd.eval(df['predicted_label'][i])
    pred_boxes = pd.eval(df['predicted_box'][i])
    name_list.append(df['img_name'][i])
    out_str = ''
    for ii in range(len(pred_labels)):
        label = pred_labels[ii]
        x0 = str(pred_boxes[ii][0]).split('.')[0]
        y0 = str(pred_boxes[ii][1]).split('.')[0]
        x1 = str(pred_boxes[ii][2]).split('.')[0]
        y1 = str(pred_boxes[ii][3]).split('.')[0]

        out_str = out_str + f'{label}:{x0},{y0},{x1},{y1} '
    
    out_str = out_str.strip(" ")
    out_list.append(out_str)
```

```python
# save list to tsv file
def save_list_to_tsv_file(path, output_list):
    with open(path, 'w', newline='') as f_output: 
        tsv_output = csv.writer(f_output, delimiter='\n')
        tsv_output.writerow(output_list)
```

```python
out_path = '../geval_input/out.tsv'
save_list_to_tsv_file(out_path, out_list)
```
